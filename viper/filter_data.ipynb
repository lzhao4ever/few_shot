### There are many small objects and owner car front in the scene which are not valid for testing our hypothesis, so we decide to remove them for the first attempt.

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
% matplotlib inline  
from PIL import Image
import numpy as np

import csv
import os
```

```python
def filter_box(box_dir, new_box_dir):
    for sub_dir in os.listdir(box_dir):
        box_files = os.listdir(os.path.join(box_dir, sub_dir))

        if not os.path.exists(os.path.join(new_box_dir, sub_dir)):
            os.mkdir(os.path.join(new_box_dir, sub_dir))
        print(os.path.join(new_box_dir, sub_dir))
        
        for box_file in box_files:
            box_file_name = os.path.join(box_dir, sub_dir, box_file)
            box = []
            with open(box_file_name) as csv_file:
                box_info = csv.reader(csv_file, delimiter=',')
                for row in box_info:
                    box.append(row)
            box = np.array(box).astype(np.float)

            anno = []
            for i in range(box.shape[0]):
                a = box[i, 2:6]
                cls = box[i, 0].astype(np.int)
                # remove owner car front
                if a[2] == 1920 and a[3] == 1080:
                    continue
                # remove small objects
                width = a[2] - a[0]
                height = a[3] - a[1]
                # box_area = (a[2] - a[0]) * (a[3] - a[1])
                # if box_area < 4096:
                #     continue
                if width < 128 or height < 128:
                    continue
                # remove non-object class
                if cls not in [19, 20, 23, 24, 25, 26, 27]:
                    continue
                anno.append([cls, a[0], a[1], a[2], a[3]])

            if len(anno) == 0:
                continue
            
            name, ext = os.path.splitext(box_file)
            new_box_file = os.path.join(new_box_dir, sub_dir, name + '.txt')
            # print(new_box_file)
            with open(new_box_file, 'w') as handle:
                for i in range(len(anno)):
                    handle.write('%d %d %d %d %d\n' % (anno[i][0], anno[i][1], anno[i][2], anno[i][3], anno[i][4]))
```

```python
train_box_dir = '/media/yi/DATA/data-orig/VIPER/train/bb'
test_box_dir = '/media/yi/DATA/data-orig/VIPER/val/bb'

train_new_box_dir = '/home/yi/code/few_shot/viper/train/bb'
test_new_box_dir = '/home/yi/code/few_shot/viper/val/bb'

filter_box(train_box_dir, train_new_box_dir)
filter_box(test_box_dir, test_new_box_dir)
```
