```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
% matplotlib inline  
from PIL import Image
import numpy as np

import csv
import os
```

```python
def compute_statistics(box_dir):
    box = []
    for sub_dir in os.listdir(box_dir):
        box_files = os.listdir(os.path.join(box_dir, sub_dir))
        print(os.path.join(box_dir, sub_dir))    
        for box_file in box_files:
            box_file_name = os.path.join(box_dir, sub_dir, box_file)
            with open(box_file_name) as txt_file:
                box_info = txt_file.readlines()
            for row in box_info:
                row = row.strip().split(' ')
                box.append(row)
    box = np.array(box).astype(np.float)
    
    cls = box[:, 0]
    box = box[:, 1:5]
    cls_labels = set(list(cls))
    count = dict()
    max_count = 0
    total_count = 0
    for cls_label in cls_labels:
        count[cls_label] = (cls == cls_label).sum()
        if count[cls_label] > max_count:
            max_count = count[cls_label]
        total_count = total_count + count[cls_label]
    print(count)
    print(max_count * 1.0 / total_count)
```

```python
train_box_dir = '/home/yi/code/few_shot/viper/train/bb'
test_box_dir = '/home/yi/code/few_shot/viper/val/bb'

compute_statistics(train_box_dir)
compute_statistics(test_box_dir)
```
